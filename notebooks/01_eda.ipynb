```python
from maternal_risk.data.load_data import load_data
from maternal_risk.data.validate import validate_schema
from maternal_risk.features.build_features import add_features

df = load_data("../data/raw/maternal_health.csv")
val = validate_schema(df)
val
```

```python
import pandas as pd

df["RiskLevel"] = df["RiskLevel"].astype(str).str.strip().str.lower()
df = add_features(df)
df.head()
```

```python
import plotly.express as px

fig = px.histogram(df, x="RiskLevel", title="RiskLevel Distribution")
fig.show()
```

```python
features = ["Age", "SystolicBP", "DiastolicBP", "BS", "BodyTemp", "HeartRate", "pulse_pressure"]

for col in features:
    fig = px.violin(df, x="RiskLevel", y=col, box=True, points="all", title=f"{col} by RiskLevel")
    fig.show()
```

```python
corr = df.drop(columns=["RiskLevel"]).corr(numeric_only=True)

fig = px.imshow(corr, title="Correlation Heatmap", aspect="auto")
fig.show()
```

```python
for col in features:
    fig = px.box(df, x="RiskLevel", y=col, title=f"Outliers: {col}")
    fig.show()
```
