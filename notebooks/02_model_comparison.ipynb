```python
import pandas as pd
import plotly.express as px

df_cmp = pd.read_csv("../reports/model_comparison.csv")
df_cmp
```

```python
fig = px.bar(df_cmp.sort_values("f1_macro", ascending=False),
             x="model_key", y="f1_macro", title="Model Macro F1 Comparison")
fig.show()
```

```python
fig = px.bar(df_cmp.sort_values("accuracy", ascending=False),
             x="model_key", y="accuracy", title="Model Accuracy Comparison")
fig.show()
```
